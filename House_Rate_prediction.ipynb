```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('C:/Users/Aaditya/Desktop/Studyagain/cal_housing_clean.csv')
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
houseage = tf.feature_column.numeric_column('housingMedianAge')
totalroom = tf.feature_column.numeric_column('totalRooms')
totalbadrom = tf.feature_column.numeric_column('totalBedrooms')
populstion = tf.feature_column.numeric_column('population')
household = tf.feature_column.numeric_column('households')
medianincom = tf.feature_column.numeric_column('medianIncome')
```

```python
feat_cols = [houseage, totalroom, totalbadrom, populstion, household, medianincom]
```

```python
x_data = dataset.drop('medianHouseValue', axis = 1)
```

```python
x_data.head()
```

```python
rat = dataset['medianHouseValue']
```

```python
rat.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, rat, test_size = 0.3, random_state = 101)
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x = X_train, y = y_train, batch_size = 32, num_epochs = 1500, shuffle = True)
```

```python
dnn_model = tf.estimator.DNNRegressor(hidden_units = [20, 30, 16, 6, 10], feature_columns = feat_cols)
```

```python
dnn_model.train(input_fn = input_func, steps = 25000)
```

```python
predict_input_func = tf.estimator.inputs.pandas_input_fn(
      x=X_test,
      batch_size=10,
      num_epochs=1,
      shuffle=False)
```

```python
pred_gen = dnn_model.predict(predict_input_func)
```

```python
predictions = list(pred_gen)
```

```python
final_preds = []
for pred in predictions:
    final_preds.append(pred['predictions'])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,final_preds)**0.5
```
